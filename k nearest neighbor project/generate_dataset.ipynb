```python
import numpy as np
import pandas as pd
```

```python
x=list(np.random.randint(0,20,100))
y=list(np.random.randint(0,20,100))
z=list(np.random.randint(0,1,100))
x1=list(np.random.randint(40,60,100))
y1=list(np.random.randint(40,60,100))
z1=list(np.random.randint(1,2,100))
x=x+x1
y=y+y1
z=z+z1
```

```python
print(x)
print(y)
print(z)
```

```python
mydict={
    "A":x,
    "B":y,
    "C":z,
}
```

```python
print(mydict)
```

```python
df=pd.DataFrame(mydict)
```
