```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv('xdata.csv')
x=x.values
x=x.reshape((99,))
y=pd.read_csv('ydata.csv')
y=y.values
y=y.reshape((99,))
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=x-x.mean()/(x.std())
X=x
Y=y
```

```python
def hipothises(theta,X):
    return theta[0]+theta[1]*X

def error(Y,X,theta):
    total_error=0.0
    m=X.shape[0]
    for i in range(m):
        total_error+=(Y[i]-hipothises(theta,X[i]))**2
    return total_error*0.5

def gradiend(Y,X,theta):
    grad=np.array([0.0,0.0])
    m=X.shape[0]
    for i in range(m):
        grad[0]+= -1.0 *(Y[i]- hipothises(theta,X[i]))
        grad[1]+= -1.0 *(Y[i]- hipothises(theta,X[i]))*X[i]
    return grad

def Gradientdicent(X,Y,learning_rate,mitr):
    grad=np.array([0.0,0.0])
    theta=np.array([0.0,0.0])
    e=[]
    
    for i in range(mitr):
        grad=gradiend(Y,X,theta)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        print(theta[0],theta[1])
        e.append(error(Y,X,theta))
       # print(error(Y,X,theta))
       # if error(Y,X,theta)<0.01:
        #    break;
        
    return theta,e
```

```python
theta,e=Gradientdicent(X,Y,0.001,100)
print(theta[0],theta[1])
```
