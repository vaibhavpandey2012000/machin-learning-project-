```python
x=pd.read_csv('xdata.csv')
x=x.values
x=x.reshape((99,))
y=pd.read_csv('ydata.csv')
y=y.values
y=y.reshape((99,))
```

```python
#what mean by nomarlization
# mean=0,stander_devation=1
x=x-x.mean()/(x.std())
plt.scatter(x,y)
plt.show()
X=x
Y=y
```

```python
#algoritham is linear regression
def hipothisis(theta,x):
    return theta[0]+theta[1]*x

def error(X,Y,theta):
    total_error=0
    m=x.shape[0]
    
    for i in range(m):
        total_error+=(Y[i]-hipothisis(theta,X))**2
    return 0.5*total_error

def gradient(Y,X,theta):
    grad=np.array([0.0,0.0])
    m=x.shape[0]
    for i in range(m):
        grad[0]+=-1.0*(Y[i]-hipothisis(theta,X[i]))
        grad[1]+=-1.0*(Y[i]-hipothisis(theta,X[i]))*X[i]
    return grad

def GradientDescet(X,Y,learning_rate,maxitr):
    grad=np.array([0.0,0.0])
    theta=np.array([0.0,0.0])
    
    for i in range(maxitr):
        grad=gradient(Y,X,theta)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
    return theta
   
```

```python
theta=GradientDescet(X,Y,learning_rate=0.001,maxitr=100)
print(theta[0],theta[1])
```
