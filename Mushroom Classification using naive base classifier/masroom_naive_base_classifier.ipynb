# Bulding Our Classifier!

```python
def prior_prob(y_train,label):
    total_ex=y_train.shape[0]
    class_ex=np.sum(y_train==label)
    return (class_ex)/float(total_ex)
```

```python
def cond_prob(x_train,y_train,feature_col,feature_val,label):
    x_fealtered=x_train[y_train==label] #its give use the all row witch crosponding colom is lable
    num=np.sum(x_fealtered[:,feature_col]==feature_val) # in all x_fealtered select those whose feature_val matched
    denom=np.sum(y_train==label)
    
    return num/float(denom)
```

# predicition

```python
def predict(x_train,y_train,xtest):
    """here xtest is single testing point ,n feature"""
    classes=np.unique(y_train)
    n_feature=x_train.shape[1]
    post_probs=[] #List of probiblity for all classes and a given single testing point
    #compute post_probs for every class
    for label in classes:
        #post_c=lickhood*prior
        likehood=1.0
        for f in range(n_feature):
            cond=cond_prob(x_train,y_train,f,xtest[f],label)
            likehood*=cond
        prior=prior_prob(y_train,label)
        post=likehood*prior
        post_probs.append(post)
    pred=np.argmax(post_probs)
    return pred

    
            
```

```python
output=predict(x_train,y_train,x_test[0])
```

```python
def score(x_train,y_train,x_test,y_test):
    pred=[]
    for i in range(x_test.shape[0]):
        pred_lable=predict(x_train,y_train,x_test[i])
        pred.append(pred_lable)
    pred=np.array(pred)
    acc=np.sum(pred==y_test)/y_test.shape[0]
    return acc
```

```python
print(score(x_train,y_train,x_test,y_test))
```
